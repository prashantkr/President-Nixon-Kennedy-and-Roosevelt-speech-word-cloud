# Speech Analysis and Word Cloud

```python
import matplotlib.pyplot as plt
%matplotlib inline
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
#Character Count
print('char count for 1941-Roosevelt.txt: ',len(inaugural.raw('1941-Roosevelt.txt')))
print('char count for 1961-Kennedy.txt: ',len(inaugural.raw('1961-Kennedy.txt')))
print('char count for 1973-Nixon.txt: ',len(inaugural.raw('1973-Nixon.txt')))
```

```python
#Word Count
print('Word count for 1941-Roosevelt.txt: ',len(inaugural.words('1941-Roosevelt.txt')))
print('Word count for 1961-Kennedy.txt: ',len(inaugural.words('1961-Kennedy.txt')))
print('Word count for 1973-Nixon.txt: ',len(inaugural.words('1973-Nixon.txt')))
```

```python
#Sentences Count
print('Sentences count for 1941-Roosevelt.txt: ',len(inaugural.sents('1941-Roosevelt.txt')))
print('Sentences count for 1961-Kennedy.txt: ',len(inaugural.sents('1961-Kennedy.txt')))
print('Sentences count for 1973-Nixon.txt: ',len(inaugural.sents('1973-Nixon.txt')))
```

```python
#Stop words
import string
stop =stopwords.words('english') + list(string.punctuation) + ['--','It','s','I','In','The','Let','let']
```

```python
roosvlt = inaugural.raw('1941-Roosevelt.txt')
kenndy = inaugural.raw('1961-Kennedy.txt')
nixon = inaugural.raw('1973-Nixon.txt')
```

```python
#Replace stop word
word_tokens = word_tokenize(roosvlt) 
filtered_roosvvlt = [w for w in word_tokens if not w in stop] 

word_tokens = word_tokenize(kenndy) 
filtered_kenndy = [w for w in word_tokens if not w in stop] 

word_tokens = word_tokenize(nixon) 
filtered_nixon = [w for w in word_tokens if not w in stop]
```

```python
#Character Count
print('Word count for 1941-Roosevelt.txt: ',len(filtered_roosvvlt))
print('Word count for 1961-Kennedy.txt: ',len(filtered_kenndy))
print('Word count for 1973-Nixon.txt: ',len(filtered_nixon))
```

```python
from collections import Counter
Counter(filtered_roosvvlt).most_common(3)
```

```python
Counter(filtered_kenndy).most_common(3)
```

```python
Counter(filtered_nixon).most_common(3)
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stop,
                min_font_size = 10).generate(roosvlt)
  
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stop,
                min_font_size = 10).generate(kenndy)
  
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```
